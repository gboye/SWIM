### Récupération des infinitifs sur Wiktionary

```python
url="https://fr.wiktionary.org/wiki/%s"
browser = mechanicalsoup.Browser()
```

```python
infinitifs={}
for l in tNeg:
    wPage=browser.get(url%l)
    wSection=wPage.soup.find("div",class_="mw-parser-output")
    if wSection:
        span=wSection.find("span",class_="API")
        if span:
            inf=span.text.strip("\\").replace(".","")
            infinitifs[l]=inf
        else:
            print l, "pas de transcription"
    else :
        print l, "pas dans Wiktionary"
print infinitifs
```
