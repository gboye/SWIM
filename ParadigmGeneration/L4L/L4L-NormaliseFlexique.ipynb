```python
import pandas as pd
import numpy as np
```

```python
repL4L="/Users/gilles/ownCloud/Recherche/Boye/HDR/Data/L4L/"
fVlexique="vlexique.csv"
```

```python
dfAPI=pd.read_csv(repL4L+fVlexique,sep=",",encoding="utf8")
dfAPI.head()
```

```python
%store -r mapCols
mapCols["variants"]="variants"
```

```python
cols=dfAPI.columns.tolist()
```

```python
apiSampa={}
apiSampa2={}
api=u"ʁəøØœɛɔɑɲʒɥʃɡŋ"
sampa=u"r6229EOaJZHSgN"
apiSampa2[u"ɑ̃"]=u"â"
apiSampa2[u"ɛ̃"]=u"ê"
apiSampa2[u"ɔ̃"]=u"ô"
apiSampa2[u"œ̃"]=u"û"
apiSampa2[u"œ̃"]=u"û"
for i,s in enumerate(api):
    apiSampa[s]=sampa[i]
```

```python
def api2sampa(api):
    result=api
    if api=="#DEF#":
        result=np.nan
    else:
        result=result.replace(";",",")
        for s in apiSampa2:
            result=result.replace(s,apiSampa2[s])
        for s in apiSampa:
            result=result.replace(s,apiSampa[s])
    return result
```

```python
def normalise(row):
    result=row
    for c in cols:
        if row[c]==row[c] and c not in ["0","lexeme","variants"]:
            result[mapCols[c]]=api2sampa(row[c])
            del result[c]
    return result
```

```python
dfSampa=dfAPI.apply(normalise,axis=1)
```

```python
dfSampa=dfSampa.drop(columns=["variants"])
```
