```python
fFlexique="vlexique.csv"
dfFlex=pd.read_csv(repL4L+fFlexique,encoding="utf8")
```

```python
cols=dfFlex.columns.tolist()
otherCols=cols[:]
otherCols.remove("variants")
flexCols=otherCols[:]
flexCols.remove("lexeme")
```

### Scinder une ligne en plusieurs avec les mêmes valeurs
https://stackoverflow.com/questions/50731229/split-cell-into-multiple-rows-in-pandas-dataframe

```python
dfF=dfFlex.set_index(otherCols).apply(lambda x: x.str.split('[;:]').explode()).reset_index()[cols]
```

```python
pbVariants=dfF[dfF.lexeme!=dfF.variants][["lexeme","variants"]].to_dict(orient="index")
```

```python
for ix in pbVariants:
    if pbVariants[ix]["variants"].startswith("sy"):
        print(ix)
```

```python
pbIndex=[621,757,760,764,765,869,873,963,990,2814,3051,3060,3082,3087,3091,3551,3574,3587,3961,4519,4522,4525,4528,4530,4533,4535,4537,4707,4924,4965,4989]
dfF[dfF.index.isin(pbIndex)][["variants","","prs.1sg"]]
```

```python
pbRepar={621:("bʁɛ̃g","bʁɛ̃k"),757:("k","Eksk"),760:("s","ʒ"),869:("siʁ","ʒiʁ"),
         873:("sit","ʒit"),963:("k","Eksk"),3051:("ʒabɔt","sabɔt"),3060:("ʒap","sap"),
         3082:("ʒut","sut"),4519:("ʁəs","s"),4522:("ʁəs","s"),4525:("ʁəs","s"),
         4528:("ʁəs","s"),4530:("ʁəs","s"),4533:("ʁəs","s"),4537:("ʁəs","s"),}
```

```python
def reparerVariants(row):
    result=row
    lexeme=row["variants"]
    variants=lexeme+":"+row["lexeme"]
    if row.name in pbRepar:
        a,b=pbRepar[row.name]
        for col in flexCols:
            result[col]=row[col].replace(a,b)
        result["lexeme"]=lexeme
        result["variants"]=variants
    return result
```

```python
dfRepar=dfF.apply(reparerVariants,axis=1)
```

```python
def dFormat(row):
    champs=[]
    champs.append(row["lexeme"])
    case=row["case"].lower()
    m=re.match(r"(.)(.)([123][sp])",case)
    if m:
        case=m.group(2)+m.group(1)+m.group(3)
    vattr="Vm%s--"%case
    champs.append(vattr[:7])
    champs.append("#%s$"%row["ortho"])
    champs.append("#%s$"%row["phono"])
    champs.append("#%s$"%row["phono"])
    champs.append("#V%s$"%row["ortho"])
    return "|".join(champs)
```

```python
df.apply(dFormat,axis=1)
```
